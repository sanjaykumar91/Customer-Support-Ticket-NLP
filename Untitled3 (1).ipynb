# Dataset Creation (Sample Database)

Since real customer support data is confidential, a sample dataset is created manually. The dataset contains a ticket ID and raw ticket text describing common customer issues such as login problems, payment failures, subscription issues, account suspension, and application crashes.

```python
import pandas as pd

data = {
    "ticket_id": [101, 102, 103, 104, 105],
    "ticket_text": [
        "I cant login to my acount since yestarday!!!",
        "Payment failed but amount debited from my bank.",
        "Need help with PayPlus subscription renewal.",
        "My account was suspended without any reason.",
        "App crashes everytime I open it on Android phone."
    ]
}

df = pd.DataFrame(data)
print(df)
```

# Text Cleaning

Text cleaning is the first step in preprocessing. Raw tickets often include uppercase letters, punctuation, numbers, and extra spaces. These elements add noise and reduce model performance.
In this step, all text is converted to lowercase, special characters are removed, and extra spaces are eliminated using regular expressions.

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["cleaned_text"] = df["ticket_text"].apply(clean_text)
```

# Spelling Correction

Customer support tickets often contain spelling mistakes due to informal typing. Spelling errors negatively affect NLP tasks such as tokenization and entity recognition.
TextBlob is used to correct spelling mistakes and improve text quality.


```python
from textblob import TextBlob

def correct_spelling(text):
    return str(TextBlob(text).correct())

df["corrected_text"] = df["cleaned_text"].apply(correct_spelling)
```

# Tokenization

Tokenization breaks text into individual words or tokens. This helps the system analyze text at the word level and enables further preprocessing steps.

spaCy is used for efficient and accurate tokenization.

#Stop-Word Removal

Stop-words are common words such as is, the, and, which do not add meaningful information. Removing stop-words reduces noise and improves data quality.

# Lemmatization

Lemmatization converts words into their base or root form while preserving meaning. This reduces vocabulary size and improves consistency across the dataset.

```python
import spacy
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    doc = nlp(text)
    tokens = []
    for token in doc:
        if token.text not in stop_words and not token.is_punct:
            tokens.append(token.lemma_)
    return " ".join(tokens)

df["final_text"] = df["corrected_text"].apply(preprocess_text)
```

# Named Entity Recognition (NER)

Named Entity Recognition identifies important entities such as product names, organizations, and platforms from the text. This helps understand the context of customer issues.



```python
def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

df["entities"] = df["corrected_text"].apply(extract_entities)
```

#  Annotation and Labeling

After preprocessing, each ticket is manually annotated with an issue category. This labeling is essential for supervised machine learning and automation.

```python
df["label"] = [
    "Login Issue",
    "Payment Issue",
    "Subscription Issue",
    "Account Suspension",
    "App Crash"
]
```

# Final Structured Dataset

**The final dataset contains:**

Ticket ID

Original text

Processed text

Extracted entities

Assigned label

This structured data is ready for classification models.

# Data Visualization (Graphs)

Graphs are used to visually validate preprocessing and annotation.

**Ticket Category Distribution Graph**

This graph shows the number of tickets in each issue category. It helps identify the most common customer problems.

```python
import matplotlib.pyplot as plt

label_counts = df["label"].value_counts()

plt.figure()
label_counts.plot(kind='bar')
plt.title("Ticket Category Distribution")
plt.xlabel("Issue Type")
plt.ylabel("Number of Tickets")
plt.show()
```

 **Word Frequency After Preprocessing**

This graph displays the most frequent meaningful words after preprocessing, proving successful noise removal.

```python
from collections import Counter

all_words = " ".join(df["final_text"]).split()
word_freq = Counter(all_words)
top_words = dict(word_freq.most_common(5))

plt.figure()
plt.bar(top_words.keys(), top_words.values())
plt.title("Top Words After Text Preprocessing")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

**Before vs After Text Length Comparison**

This graph compares text length before and after preprocessing, showing the reduction of unnecessary data.

```python
before_length = df["ticket_text"].apply(len)
after_length = df["final_text"].apply(len)

plt.figure()
plt.bar(["Before Cleaning", "After Cleaning"],
        [before_length.sum(), after_length.sum()])
plt.title("Text Length Comparison")
plt.ylabel("Total Text Length")
plt.show()
```
